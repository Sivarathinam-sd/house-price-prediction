# DATA CLEANING

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df=df.drop(['priceSqFt','verificationDate','description'],axis=1)
```

```python
df.info()
```

```python
df['numBathrooms']=df['numBathrooms'].fillna(0.0)
df['numBathrooms'].unique()
```

```python
df['numBalconies']=df['numBalconies'].fillna(0.0)
df['numBalconies'].unique()
```

```python
df['isNegotiable']=df['isNegotiable'].fillna("Not Negotiable")
df['isNegotiable'].unique()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
def extract_house_size(house_size): #utility function
    price= house_size.split(" ")
    price = price[0].split(",")
    price = "".join(price)
    return float(price)

df['converted_house_size'] = df['house_size'].apply(lambda x: extract_house_size(x))
df['converted_house_size'].unique()
```

```python
df['price']=df['price'].astype(float)
df['price'].dtype
```

```python
df['currency'].unique()
```

```python
df['isNegotiable'].unique()
```

```python
def clean_security_deposit(security_deposit):
    if security_deposit == 'No Deposit':
        return 0.0
    else:
        security_deposit = security_deposit.split(",")
        security_deposit = "".join(security_deposit)
        security_deposit.strip()
        return int(security_deposit)
    
df['cleaned_security_deposit'] = df['SecurityDeposit'].apply(lambda x: clean_security_deposit(x))
df['cleaned_security_deposit'].dtype
```

```python
df['Status'].unique()
```

```python
df.info()
```

```python
df=df.drop(['SecurityDeposit', 'house_size'],axis=1)
df.head()
```

```python
def get_numRooms(house_type):
    return int(house_type.split(" ")[0])

def get_configuration(house_type):
    return house_type.split(" ")[1]

def get_house_type(house_type):
    house_type = " ".join(house_type.split(" ")[2:])
    return house_type.strip()

df['numRooms'] = df['house_type'].apply(lambda x: get_numRooms(x))
df['house_configuration'] = df['house_type'].apply(lambda x: get_configuration(x))
df['cleaned_house_type'] = df['house_type'].apply(lambda x: get_house_type(x))
df.info()
```

```python
df = df.drop('house_type', axis=1)
df.head()
```

```python
df.columns
```

```python
df_pune = df[['numRooms', 'house_configuration', 'cleaned_house_type', 
               'location', 'city', 'latitude', 'longitude', 'currency',
                'numBathrooms', 'numBalconies', 'isNegotiable', 'Status',
               'converted_house_size', 'cleaned_security_deposit', 'price']]

df_pune.to_csv('../../datasets/cleaned_pune_data.csv')
```
